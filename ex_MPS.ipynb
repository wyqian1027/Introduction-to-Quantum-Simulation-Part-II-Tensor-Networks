### c. MPS by hand

For example, the generalized GHZ state is
$$\begin{align}
    \ket{GHZ}=\frac{\ket{00...0}+\ket{11...1}}{\sqrt{2}}\,,
\end{align}$$
which can be constructed with open boundary using
$$\begin{align}
    A^{(1)}_0 &= \begin{pmatrix}
        1/\sqrt{2} & 0
    \end{pmatrix}\,,\quad\quad
    A^{(1)}_1 = \begin{pmatrix}
        0 & 1/\sqrt{2}
    \end{pmatrix}\,,\\
    A^{(k)}_0 &= \begin{pmatrix}
        1 & 0\\
        0 & 0
    \end{pmatrix}\,,\quad
    A^{(k)}_1 = \begin{pmatrix}
        0 & 0\\
        0 & 1
    \end{pmatrix}\,,\\
    A^{(N)}_0 &= \begin{pmatrix}
        1\\0
    \end{pmatrix}\,,\quad\quad
    A^{(N)}_1 = \begin{pmatrix}
        0\\ 1
    \end{pmatrix}\,,
\end{align}$$
where $k=2,3,...,N-1$.

```julia
N = 5

sites = siteinds("Qubit", N)

# Create all link indices first
links = [Index(2, "Link,l=$i") for i in 1:N-1]

# Create tensor by hand

# Site 1
A1 = ITensor(sites[1], links[1])
A1[sites[1] => 1, links[1] => 1] = 1/√2  # |0⟩ → bond 1
A1[sites[1] => 2, links[1] => 2] = 1/√2  # |1⟩ → bond 2
tensors = [A1]

# Middle sites
for i in 2:N-1
    A = ITensor(sites[i], links[i-1], links[i])
    A[sites[i] => 1, links[i-1] => 1, links[i] => 1] = 1.0
    A[sites[i] => 2, links[i-1] => 2, links[i] => 2] = 1.0
    push!(tensors, A)
end

# Last site
AN = ITensor(sites[N], links[N-1])
AN[sites[N] => 1, links[N-1] => 1] = 1.0
AN[sites[N] => 2, links[N-1] => 2] = 1.0
push!(tensors, AN)

# Create MPS from tensors
ghz = MPS(tensors)
```

```julia
function make_ghz_mps(sites)
    
    N = length(sites)
    
    # Create all link indices first
    links = [Index(2, "Link,l=$i") for i in 1:N-1]
    
    # Create tensor by hand
    
    # Site 1
    A1 = ITensor(sites[1], links[1])
    A1[sites[1] => 1, links[1] => 1] = 1/√2  # |0⟩ → bond 1
    A1[sites[1] => 2, links[1] => 2] = 1/√2  # |1⟩ → bond 2
    tensors = [A1]
    
    # Middle sites
    for i in 2:N-1
        A = ITensor(sites[i], links[i-1], links[i])
        A[sites[i] => 1, links[i-1] => 1, links[i] => 1] = 1.0
        A[sites[i] => 2, links[i-1] => 2, links[i] => 2] = 1.0
        push!(tensors, A)
    end
    
    # Last site
    AN = ITensor(sites[N], links[N-1])
    AN[sites[N] => 1, links[N-1] => 1] = 1.0
    AN[sites[N] => 2, links[N-1] => 2] = 1.0
    push!(tensors, AN)
    
    # Create MPS from tensors
    ghz = MPS(tensors)
    return ghz
    
end
    
```

## 2. Matrix Product Operator

```julia
N = 5

sites = siteinds("Qubit", N)

os = OpSum()
os .+= "Z", 5

MPO(os, sites)
```

```julia
function make_X_mpo(sites, i)
    os = OpSum()
    os .+= "X", i          # Add Pauli X on site i
    return MPO(os, sites)
end

function make_Z_mpo(sites, i)
    os = OpSum()
    os .+= "Z", i          # Add Pauli Z on site i
    return MPO(os, sites)
end

# correlation/parity
function make_ZZ_mpo(sites, i, j)
    os = OpSum()
    os .+= "Z", i, "Z", j  # Add Z_i * Z_j term
    return MPO(os, sites)
end
```

```julia
ghz = make_ghz_mps(sites)

for i in 1:N
    println("< X$i > = ", inner(ghz', make_X_mpo(sites, i), ghz))
end
```

```julia
ghz = make_ghz_mps(sites)

for i in 1:N
    println("< Z$i > = ", inner(ghz', make_Z_mpo(sites, i), ghz))
end
```

```julia
ghz = make_ghz_mps(sites)

for i in 1:N
    for j in 1:N
        println("< Z$i Z$j > = ", inner(ghz', make_ZZ_mpo(sites, i, j), ghz))
    end
end
```

## 3. Entanglement entropy

```julia
# entropy of entanglement can be obtained from SVD

function entropy_bond(psi)
    N = length(psi)
    S_values = zeros(N-1)
    for n in 1:N-1
        # put orthogonality center on bond n
        orthogonalize!(psi, n)
        # SVD of site n tensor
        A = psi[n]
        U, S, V = svd(A, (siteinds(psi, n)))
        s = storage(S)
        p = s.^2
        p = p[p .> 1e-14]   # remove numerical zeros
        S_values[n] = -sum(p .* log2.(p))
    end
    return S_values
end


function entropy_2_sites(psi) 
    N = length(psi)
    S_values = zeros(N-1)
    for n in 1:N-1
        orthogonalize!(psi, n)
        if n == 1
            Linds = (siteinds(psi, n), siteinds(psi, n+1))
        else
            Linds = (linkind(psi, n), siteinds(psi, n), siteinds(psi, n+1))
        end
        _, S, _ = svd(psi[n] * psi[n+1], Linds)
        Svals = storage(S)
        p_values = Svals .^ 2
        p_values = filter(x -> x > 0, p_values)
        S_values[n] = -sum(p_values .* log2.(p_values))
    end
    return S_values
end
```

```julia
N = 10

sites = siteinds("S=1/2", N)

states = [isodd(n) ? "Up" : "Dn" for n=1:N]

psi = productMPS(sites, states)

linkdims = [dim(linkind(rand_psi, i)) for i in 1:N-1]

println(linkdims)

entropy_bond(psi)
# entropy_2_sites(psi)
```
